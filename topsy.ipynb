```python
dia, l, t, factor = 50, 100, 5, 2.4

with BuildPart() as topsy:
    with BuildSketch() as topsy_sk:
        with BuildLine() as topsy_ln:
            l1 = Line((0,-dia/2), (dia/factor,0))
            round = RadiusArc((dia/factor,0),(-dia/factor,0),dia/2.0,short_sagitta=False)
            l2 = Line(round @ 1, l1 @ 0)
        hole = make_face()
        offset(hole, amount=t/2, kind=Kind.TANGENT)
        # Rectangle(t, l, align=(Align.CENTER,Align.MAX))
        Trapezoid(width=t, height=l, left_side_angle=91, align=(Align.CENTER,Align.MAX))
        add(hole, mode=Mode.SUBTRACT)
    extrude(amount=t)
    chamfer(topsy.edges().group_by(Axis.Y)[0].filter_by(Axis.X).sort_by(Axis.Z)[1], length=t/1.3, angle=80)
    fillet(topsy.edges().group_by(Axis.Y)[4:], radius=t/5)
    # RigidJoint(label="middle", joint_location=Location(round.center()-(0,dia/2,0)))
    RigidJoint(label="middle", joint_location=Location(round.edge().arc_center))

topsy.part
```
